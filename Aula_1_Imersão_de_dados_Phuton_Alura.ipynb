```python
columns_translation = {
    'ano': 'ano',
    'nivel_experiencia': 'senioridade',
    'tipo_emprego': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'eresidencia_funcionario': 'residencia',
    'taxa_remota': 'remoto',
    'local_empresa': 'empresa',
    'tamanho_empresa': 'tamanho_empresa'
}
df = df.rename(columns=columns_translation)
print("Colunas traduzidas com sucesso!")
print(df.columns)
```

```python
columns_translation_corrected = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}
df = df.rename(columns=columns_translation_corrected)
df["senioridade"].value_counts()
```

```python
df["contrato"].value_counts()
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python


senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df["senioridade"] = df["senioridade"].replace(senioridade)
df["senioridade"].value_counts()
```

```python


contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(contrato)

print("Categorias da coluna 'contrato' traduzidas com sucesso!")
print(df['contrato'].value_counts())
```

```python
tamanho_empresa_translation = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa_translation)

print("Categorias da coluna 'tamanho_empresa' traduzidas com sucesso!")
print(df['tamanho_empresa'].value_counts())
```

```python
remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(remoto)

print("Categorias da coluna 'remoto' traduzidas com sucesso!")
print(df['remoto'].value_counts())
```

```python
import numpy as np

# Criação de um dataframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
     "nome":['ana', ' Bruno' , 'Carlos' , 'Daniele' , 'Val'] ,
     'salario': [4000, np.nan, 5000, np.nan , 100000]
})

# calcula a média salarial e subistitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))


# Calcula mediana e subistitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
import plotly.express as px
```

```python
df_senioridade_usd = df_limpo.groupby('senioridade')['usd'].mean().reset_index()

ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

fig = px.bar(df_senioridade_usd,
             x='senioridade',
             y='usd',
             title='Média Salarial por Senioridade (USD)',
             labels={'senioridade': 'Senioridade', 'usd': 'Média Salarial Anual (USD)'},
             category_orders={'senioridade': ordem_senioridade})

fig.update_layout(xaxis_title='Senioridade', yaxis_title='Média Salarial Anual (USD)')
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
            names='tipo_trabalho',
            values='quantidade',
            title='Proporção dos tipos de trabalho'
)

fig.show()
```
